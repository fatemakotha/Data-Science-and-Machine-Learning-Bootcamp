```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('E:\COMPLETE 2022 Data Science & Machine Learning Bootcamp\SECTION 2 Predict Movie Box Office Revenue with Linear Regression\cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
3.29e7
```

***CREATE NEW DATAFRAMES:***

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
print(type(X))
print(type(y))
```

***CODE TO RUN LINEAR REGRESSION***

```python
#Create a new LinearRegression object:
regression = LinearRegression()
#Tell the object to run a regression:
regression.fit(X, y) #X=features, y=labels
```

***Get hold of the SLOPE (coefficient):***

```python
#Get hold of the SLOPE:
regression.coef_ #theta_1
```

***Get the intercept of the regression line***

```python
#Intercept
regression.intercept_
```

***Plot the regression line on the scatter chart:***

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
#Plot the regression line:
plt.plot(X, regression.predict(X)) #X is the production_budget_usd, and y is the predicted value of X ***


plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```
